```python
# Student Marks Analyzer

print("=== Student Marks Analyzer ===")

# Get number of students
n = int(input("Enter number of students: "))

# Create an empty dictionary to store student names and marks
marks = {}

# Input student names and marks
for i in range(n):
    name = input(f"\nEnter name of student {i+1}: ")
    score = float(input(f"Enter marks of {name}: "))
    marks[name] = score

# Calculate statistics
total = sum(marks.values())
average = total / n
highest = max(marks.values())
lowest = min(marks.values())

# Find students with highest and lowest marks
topper = [name for name, score in marks.items() if score == highest]
bottom = [name for name, score in marks.items() if score == lowest]

# Display results
print("\n--- Results ---")
print(f"Average Marks: {average:.2f}")
print(f"Highest Marks: {highest} ({', '.join(topper)})")
print(f"Lowest Marks: {lowest} ({', '.join(bottom)})")

# Grade assignment
print("\n--- Grades ---")
for name, score in marks.items():
    if score >= 90:
        grade = 'A+'
    elif score >= 80:
        grade = 'A'
    elif score >= 70:
        grade = 'B'
    elif score >= 60:
        grade = 'C'
    else:
        grade = 'D'
    print(f"{name}: {score} â†’ Grade {grade}")
```
