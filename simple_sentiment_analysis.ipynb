```python
from textblob import TextBlob
text1 = "i love this product & it's amazing"
blob1 = TextBlob(text1)
sentiment1 = blob1.sentiment.polarity
print(sentiment1)
```

```python
print("=" * 60)
print("TEXT 1:", text1)
print(f"Sentiment Score: {sentiment1:.2f}")
```

```python
if sentiment1 > 0:
    print("The sentiment is positiveüòÅ.")
elif sentiment1 < 0:
    print("The sentiment is negativeü§¢.")
else:
    print("The sentiment is neutralüòä.")
```
